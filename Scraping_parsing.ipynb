```python
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
import numpy as np
import time
```

```python
### This code makes openning browser in background
options = Options()
options.binary_location = "/usr/bin/firefox"
options.add_argument("-headless")

### Initiating browser
driver = webdriver.Firefox(options=options)

### URL
url = "https://auto.ria.com/uk/search/?categories.main.id=1&brand.id[0]=76&year[0].gte=2000&indexName=auto,order_auto,newauto_search&body.id[0]=307&body.id[1]=449&body.id[2]=8&body.id[3]=2&body.id[4]=5&body.id[5]=3&body.id[6]=4&credit=0&confiscated=0&size=20&page=40&size=20"


# url = "https://auto.ria.com/uk/search/?indexName=auto,order_auto,newauto_search&categories.main.id=1&year%5B0%5D.gte=2012&verified.VIN=1&plateNumber.length.gte=1&body.id%5B0%5D=3&price.USD.gte=9000&credit=0&confiscated=0&fuel.id%5B0%5D=2&gearbox.id%5B0%5D=2&gearbox.id%5B1%5D=3&gearbox.id%5B2%5D=4&gearbox.id%5B3%5D=5&drive.id%5B0%5D=2&fuel.consumption.city.gte=2&fuel.consumption.city.lte=12&engine.lte=2.5&mileage.lte=300"
### Calling browser
driver.get(url)

### Getting html code of site
s = requests.Session()
# Set correct user agent
selenium_user_agent = driver.execute_script("return navigator.userAgent;")
s.headers.update({"user-agent": selenium_user_agent})

for cookie in driver.get_cookies():
    s.cookies.set(cookie['name'], cookie['value'], domain=cookie['domain'])

```

```python
first_text = r.text
for lead in _li:
    # TYPE
    soup = BeautifulSoup(first_text, 'html.parser')
    ttt = soup.find_all('img', class_="outline m-auto")
    print(ttt[0]['alt'])

    url1 = lead['href']
    r = s.get(url1)

    # PRICE
    soup = BeautifulSoup(r.text, 'html.parser')
    li = soup.find_all('span', class_="price_value")
    print(''.join(re.findall(r'[\d ]+', str(li[0]))).strip())

    # MODEL and YEAR
    li = soup.find_all('div', class_="ticket-status-0")
    sub_li = soup.find_all('h1', class_="head")
    print(sub_li[0]['title'])

    # SOURCE AND DTP
    li = soup.find_all('ul', class_="unstyle label-param")
    if li:
        new_soup = BeautifulSoup(str(li[0]), 'html.parser')
        new_li = new_soup.find_all('li', class_="item")
        for tag in new_li:
            print(tag.text)

    # CHECKED BY AUTORIA
    li = soup.find_all('div', class_="technical-info ticket-checked")
    if li:
        new_soup = BeautifulSoup(str(li[0]), 'html.parser')
        new_li = new_soup.find_all('span', class_="label")
        for tag in new_li:
            print(tag.text)
        new_li = new_soup.find_all('span', class_="argument")
        for tag in new_li:
            print(tag.text)

    # PAID CHECKED BY AUTORIA
    li = soup.find_all('div', class_="paid technical-info ticket-checked")
    if li:
        new_soup = BeautifulSoup(str(li[0]), 'html.parser')
        new_li = new_soup.find_all('span', class_="label")
        for tag in new_li:
            print(tag.text)
        new_li = new_soup.find_all('span', class_="argument")
        for tag in new_li:
            print(tag.text)

    # DESCRIPTION INFO
    li = soup.find_all('div', class_="box-panel description-car")
    if li:
        new_soup = BeautifulSoup(str(li[0]), 'html.parser')
        new_li = new_soup.find_all('span', class_="label")
        for tag in new_li:
            print(tag.text)
        new_li = new_soup.find_all('span', class_="argument")
        for tag in new_li:
            print(tag.text)

    # FULL DESCRIPTION
    soup = BeautifulSoup(r.text, 'html.parser')
    li = soup.find_all('div', class_="full-description")
    print(str(li[0]))




    print("\n\n")
```

```python
### This code get reviews from site and save them into .csv
tic = time.perf_counter()

i=1
reviews = []
r = s.get(url)
while r.status_code == 200 and i<5:# and i<5
    ### Initiating soup object
    soup = BeautifulSoup(r.text, 'html.parser') ### https://medium.com/analytics-vidhya/word2vector-using-gensim-e055d35f1cb4
    _li = soup.find_all('div', class_='result-explore fl-r m-view')
    reviews.extend(_li)
    ### Getting list of reviews from site
    r = s.get(url + f'&page={i}&size=50')
    print(i, r, url + f'&page={i}&size=50')
    i += 1

    
toc = time.perf_counter()
print(f"Time to get all reviews {(toc - tic)/60} minutes")


df = pd.DataFrame({"reviews":reviews})
# df.reviews = df.reviews.str.replace('&amp;', "")

df.to_csv('Reviews.csv', index=False)
```

```python
### Code below add to df columns with dates, etc.

def get_date(row):
    soup = BeautifulSoup(str(row), 'html.parser')
    try:
        date = re.search(r'[a-zA-Z]+ [0-9]{1,2}, [0-9]{4}', str(soup.find_all('div', class_='css-8a5o2x e1wnkr790'))).group(0).replace(">", "").replace("<", "")
    except AttributeError:
        date = pd.NA
    del soup
    return date

def get_topic(row):
    soup = BeautifulSoup(str(row), 'html.parser')
    topic = re.search(r'>(.*)<', str(soup.find_all('h2', class_='css-wn2j4x e1tiznh50')[0].find_all('span', class_='css-82l4gy eu4oa1w0')[0])).group(0)
    return topic.replace(">", "").replace("<", "")

def get_raiting_value(row):
    soup = BeautifulSoup(str(row), 'html.parser')
    raiting_value = re.search(r'>(.*)<', str(soup.find_all('button', class_='css-1c33izo e1wnkr790'))).group(0)
    return raiting_value.replace(">", "").replace("<", "")
    
def get_body(row):
    soup = BeautifulSoup(str(row), 'html.parser')
    body = soup.find_all('span', class_='css-1cxc9zk e1wnkr790')[0].find_all('span', class_='css-82l4gy eu4oa1w0')
    full_body = ''
    for b in body:
        try:
            full_body = full_body + re.search(r'>[^<](.*)<', str(b)).group(0)
        except AttributeError:
            full_body = full_body
    return full_body.replace(">", "").replace("<", "")

def get_pros(row):
    soup = BeautifulSoup(str(row), 'html.parser')
    try:
        pros = re.search(r'>(.*)<', str(soup.find_all('div', class_='css-rr5fiy eu4oa1w0')[1].find_all('span', class_='css-82l4gy eu4oa1w0')[0])).group(0).replace(">", "").replace("<", "")
    except IndexError:
        pros = pd.NA
    return pros

def get_cons(row):
    soup = BeautifulSoup(str(row), 'html.parser')
    try:
        cons = re.search(r'>(.*)<', str(soup.find_all('div', class_='css-rr5fiy eu4oa1w0')[1].find_all('span', class_='css-82l4gy eu4oa1w0')[1])).group(0).replace(">", "").replace("<", "")
    except IndexError:
        cons = pd.NA
    return cons




tic = time.perf_counter()        


df["date"] = df["reviews"].apply(get_date)  
df["topic"] = df["reviews"].apply(get_topic)  
df["raiting_value"] = df["reviews"].apply(get_raiting_value)  
df["body"] = df["reviews"].apply(get_body)  
df["pros"] = df["reviews"].apply(get_pros)  
df["cons"] = df["reviews"].apply(get_cons)  


toc = time.perf_counter()
print(f"Time to get all reviews {(toc - tic)/60} minutes")

df.to_csv('Reviews.csv', index=False)
```
